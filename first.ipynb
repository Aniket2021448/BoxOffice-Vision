```python
# # Other pre processing steps, 
# 1. remove the dollar sign in Budget and conver the column to numerical field.
# 2. conver the release date to datetime field.
# 3. Convert the runtime to time field(minutes).
# 4. Convert rating count to numerical field. 
# 5. make bins of the rating field to create the target variable field of HIT, AVERAGE, FLOP.
# 6. remove the unnamed field from the begining of the dataset.

# and name the final dataset as movies_cleaned.csv
```

```python
movies['Budget'] = movies['Budget'].replace({'\$': '', '€': '', '£': '', ',': ''}, regex=True).astype(float)
movies.head()
```

```python
movies['Release_Data'] = pd.to_datetime(movies['Release_Data'])
```

```python
movies.info()
```

```python
def convert_runtime(runtime):
    hours = 0
    minutes = 0
    
    # Use regular expressions to capture hours and minutes
    hour_match = re.search(r'(\d+)\s*hour', runtime)
    minute_match = re.search(r'(\d+)\s*minute', runtime)
    
    # If an hour part is found, convert it to minutes
    if hour_match:
        hours = int(hour_match.group(1))
        
    # If a minute part is found, add the minutes
    if minute_match:
        minutes = int(minute_match.group(1))
    
    # Convert hours to minutes and return total
    return hours * 60 + minutes

# Apply the conversion function to the 'Runtime' column
movies['Runtime'] = movies['Runtime'].apply(convert_runtime)
```

```python
bins = [0, 5, 7, 10]  # FLOP: 0-5, AVERAGE: 5-7, HIT: 7-10
labels = ['FLOP', 'AVERAGE', 'HIT']
movies['Target'] = pd.cut(movies['Rating'], bins=bins, labels=labels)
```

```python
movies = movies.loc[:, ~movies.columns.str.contains('^Unnamed')]
```

```python
def convert_rating_count(value):
    if 'K' in value:
        return float(value.replace('K', '')) * 1_000
    elif 'M' in value:
        return float(value.replace('M', '')) * 1_000_000
    else:
        return float(value)

movies['Rating_Count'] = movies['Rating_Count'].apply(convert_rating_count)
```
